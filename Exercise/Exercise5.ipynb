```python
import math # 要使⽤圓周率 pi 得引入套件 math
def circle_calculator(r):
 '''
 輸入半徑長度可以計算出圓⾯積與圓周長
 將圓⾯積與圓周長⼀起回傳
 '''
 area = math.pi * r ** 2
 circum = 2 * math.pi * r
 return area, circum
print (circle_calculator(3))
```
